```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_gamma_85 = pd.read_csv("sac_progress_gamma_85.csv")
data_gamma_90 = pd.read_csv("sac_progress_gamma_90.csv")
data_gamma_95 = pd.read_csv("sac_progress_gamma_95.csv")
```

```python
plt.plot(data_gamma_85["time/total_timesteps"], data_gamma_85["rollout/ep_rew_mean"], label="gamma = 0.85")
plt.plot(data_gamma_90["time/total_timesteps"], data_gamma_90["rollout/ep_rew_mean"], label="gamma = 0.90")
plt.plot(data_gamma_95["time/total_timesteps"], data_gamma_95["rollout/ep_rew_mean"], label="gamma = 0.95")
plt.xlabel("Steps", fontsize=12)
plt.ylabel("Reward", fontsize=12)
plt.title("Soft Actor-Critic reward with different gamma", fontsize=12)
plt.legend()
```

```python
plt.plot(data_gamma_85["time/total_timesteps"], data_gamma_85["train/ent_coef"], label="gamma = 0.85")
plt.plot(data_gamma_90["time/total_timesteps"], data_gamma_90["train/ent_coef"], label="gamma = 0.90")
plt.plot(data_gamma_95["time/total_timesteps"], data_gamma_95["train/ent_coef"], label="gamma = 0.95")
plt.xlabel("Steps", fontsize=12)
plt.ylabel("Entropy coefficient", fontsize=12)
plt.title("Soft Actor-Critic entropy", fontsize=12)
plt.legend()
```
